Amazon Product Web Scrapping 

In this Project I have Collect Tshirt data in 3 ways:
1. Single Prodcut Info
2. Single Page Product Info
3. Multi Page Product Info

In Single Product Scrapping I have also added a sleeper so that we can get the updated data in a particular interval of time.

I have also added a send_mail function so that if we want to get notified when  a particular product get below our given certain value.

Scrapping A Single Product from Amazon

```python
# Connect to Website

URL = 'https://www.amazon.in/DUDEME-Programmer-Coding-Developer-T-Shirt/dp/B08SFHKHV3/ref=sr_1_1?crid=1Y8XLS4WLKXLJ&keywords=data%2Banalyst%2Btshirt&qid=1673178175&sprefix=data%2Banalyst%2Btshirt%2Caps%2C410&sr=8-1'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64; rv:108.0) Gecko/20100101 Firefox/108.0"}

page = requests.get(URL , headers=headers)
soup1 = BeautifulSoup(page.content, "html.parser")
soup2= BeautifulSoup(soup1.prettify(),"html.parser")

title = soup2.find(id="productTitle").get_text()
price = soup2.find(attrs={'class':"a-offscreen"}).get_text()


print(price)
```

```python
#TO CLEAN THE WHITE SPACE AROUND PRICE AND REMOVE THE RUPPEE SIGN
price = price.strip()[1:]
title =title.strip()
print(title)
print(price)
```

```python
today = datetime.date.today()
print(today)
```

Scrapping A Page Product Info from Amazon

```python
title= []
price= []
URL = 'https://www.amazon.in/s?k=data%2Banalyst%2Btshirt&ref=nb_sb_noss'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64; rv:108.0) Gecko/20100101 Firefox/108.0"}

page = requests.get(URL , headers=headers)
soup1 = BeautifulSoup(page.content, "html.parser")
soup2= BeautifulSoup(soup1.prettify(),"html.parser")

for item in soup2.find_all("div", class_="a-section a-spacing-none a-spacing-top-small s-title-instructions-style"):
        title.append(item.h2.text)


for item in soup2.find_all("span", class_="a-price-whole"):
        price.append(item.text)




#title = soup2.findall(id="productTitle").get_text()
#price = soup2.find(attrs={'class':"a-offscreen"}).get_text()
```

```python
#To Remove extra space and irrelevant Items
price =[j.strip('\n                         \n') for j in price]
price = [j.lstrip('') for j in price]
print(price)
```

```python
#To Remove extra space and irrelevant Items
title =[i.strip('\n') for i in title]
titles = [i.strip() for i in title]
```

Scrapping Multi Page Data

```python
titlepg= []
pricepg= []
for i in range(1,6):
   URL = f'https://www.amazon.in/s?k=t-shirt&ref=nb_sb_noss/page/{i}/?sortby=post_date%3ADesc&pagesize=100'
   headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64; rv:108.0) Gecko/20100101 Firefox/108.0"}

   page = requests.get(URL , headers=headers)
   soup1 = BeautifulSoup(page.content, "html.parser")
   soup2= BeautifulSoup(soup1.prettify(),"html.parser")

   for item in soup2.find_all("span", class_="a-size-base-plus a-color-base a-text-normal"):
        titlepg.append(item.text)
        print('scrapping done')


   for item in soup2.find_all("span", class_="a-price-whole"):
        pricepg.append(item.text)
```
